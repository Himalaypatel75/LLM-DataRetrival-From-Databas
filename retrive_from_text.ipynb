```python
text = []

with open('data.txt', 'r') as file:
    for line in file:
        # Remove leading and trailing whitespaces and split by comma
        values = line.strip().split(',')
        text = values
text
```

```python
from langchain.chains import LLMChain, StuffDocumentsChain
from langchain.prompts import PromptTemplate
from langchain_community.document_transformers import (
    LongContextReorder,
)
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

# Get embeddings.
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
retriever = Chroma.from_texts(text, embedding=embeddings).as_retriever(
    search_kwargs={"k": 30}
)
query = "Can you give me related suggession for house wife?"

# Get relevant documents ordered by relevance score
docs = retriever.get_relevant_documents(query)
required_products1 = set()
for i in docs:
    required_products1.add(i.page_content)
print(required_products1)

print("--")
query1 = "Can you give me related suggession for student?"

# Get relevant documents ordered by relevance score
docs1 = retriever.get_relevant_documents(query1)
required_products2 = set()
for i in docs1:
    required_products2.add(i.page_content)
print(required_products2)


print("--")
query2 = "Can you give me related suggession for car mechanic?"

# Get relevant documents ordered by relevance score
docs2 = retriever.get_relevant_documents(query2)
required_products3 = set()
for i in docs2:
    required_products3.add(i.page_content)
print(required_products3)
```
